## Toy MC of Crystal Ball fit of Convolution

```python
def crystalball_pdf(x, alpha, n, xbar, sigma):
    n_over_alpha = n/abs(alpha)
    exp = np.exp(-0.5*alpha ** 2)
    A = (n_over_alpha)**n*exp
    B =  n_over_alpha - abs(alpha)
    C = n_over_alpha/(n-1)*exp
    D = np.sqrt(0.5*np.pi)*(1 + scipy.special.erf(abs(alpha)/np.sqrt(2)))
    N = 1/(sigma*(C + D))

    mask = (x - xbar)/sigma > -alpha

    gaussian = N*np.exp(-0.5*((x[mask]-xbar)/sigma)**2)
    powerlaw = N*A*(B - (x[~mask]-xbar)/sigma)**-n

    result = np.zeros_like(x)
    result[mask] = gaussian
    result[~mask] = powerlaw
    
    return result
```

```python
def chi2(N,w,alpha, n,mu,sig, sig2):
    return sum(
        (N*w*crystalball_pdf(x, alpha, n, mu, sig)+N*(1-w)*scipy.stats.norm.pdf(x,mu, sig2) -  dist)**2 / dist**2)
```

```python
results = []
for sigma in np.linspace(1,10,50)[1:]:
    x=np.linspace(-7*sigma,7*sigma,500)
    dist = tanh_dp._pdf(x,1,0,sigma)
    min = iminuit.Minuit(chi2,1,.5,1,3,0,sigma,sigma)
    min.limits = [
        (0,np.inf),
        (0,1),
        (0,np.inf),
        (0,np.inf),
        (-np.inf,np.inf),
        (0,np.inf),
        (0,np.inf),
    ]
    min.migrad()
    #print(sigma)
    results.append(min)
```

```python
x=np.linspace(1,10,50)[1:]
alphas = [res.values['alpha'] for res in results]
ns = [res.values['n'] for res in results]
ws = [res.values['w'] for res in results]
werrs = [res.errors['w'] for res in results]
alphaerrs = [res.errors['alpha'] for res in results]
nerrs = [res.errors['n'] for res in results]
mus = [res.values['mu'] for res in results]
muerrs = [res.errors['mu'] for res in results]
sig1s = [res.values['sig'] for res in results]
sig1errs = [res.errors['sig'] for res in results]
sig2s = [res.values['sig2'] for res in results]
sig2errs = [res.errors['sig2'] for res in results]
muosig1 = np.array(mus)/np.array(sig1s)
```

```python
def sample():
    raw_data_dp = np.random.choice(Dp_mother_mass, size=800000) + np.multiply(np.random.choice(Dp_DTF_MERR,800000),np.random.randn(800000)) 
    raw_data_ds = np.random.choice(Ds_mother_mass, size=1000000) + np.multiply(np.random.choice(Dp_DTF_MERR,1000000),np.random.randn(1000000))
    background = np.random.exponential(150, 60000)+1840
    return np.concatenate([raw_data_dp,raw_data_ds,background])
```
